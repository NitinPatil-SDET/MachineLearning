# Why Do We Need Different Types of ML?

Not all datasets are the same.

Sometimes we know the correct answer.

Sometimes we don't.

Sometimes we know a few answers.

Sometimes the machine learns through rewards.

Because of these different situations, Machine Learning is divided into multiple categories.

---

# Types of Machine Learning

```text
Machine Learning
│
├── Supervised Learning
│
├── Unsupervised Learning
│
├── Semi-Supervised Learning
│
└── Reinforcement Learning
```

---

# 1. Supervised Learning

## Definition

In Supervised Learning, the dataset contains:

* Input Features (X)
* Correct Output (Y)

The model learns the relationship between X and Y.

---

## Example

House Price Prediction

| Area | Price   |
| ---- | ------- |
| 1000 | 50 Lakh |
| 1500 | 70 Lakh |
| 2000 | 95 Lakh |

Here:

```text
Area = Input
Price = Output
```

The model learns:

```text
Area → Price
```

and predicts future prices.

---

## Key Idea

```text
Input + Correct Answer
```

The model learns from labeled data.

---

# What is Labelled Data?

Labelled Data means:

```text
Input + Output
```

Example:

| Email   | Spam |
| ------- | ---- |
| Email A | Yes  |
| Email B | No   |
| Email C | Yes  |

The output column is called the label.

---

# Types of Supervised Learning

## 1. Regression

Used when output is continuous.

Examples:

* House Price Prediction
* Salary Prediction
* Temperature Prediction

Output:

```text
25.5
70.2
100.8
```

Numbers.

---

## 2. Classification

Used when output belongs to categories.

Examples:

* Spam / Not Spam
* Pass / Fail
* Fraud / Not Fraud
* Disease / No Disease

Output:

```text
Yes
No
Pass
Fail
```

Categories.

---

# Common Algorithms

Regression:

* Linear Regression
* Polynomial Regression

Classification:

* Logistic Regression
* Decision Tree
* Random Forest
* SVM
* KNN

---

# Real-Life Examples

### Gmail

Spam Detection

### Banks

Fraud Detection

### Netflix

Movie Recommendation (partially supervised)

### Insurance

Claim Risk Prediction

---

# Advantages

* High accuracy when labels are available
* Easy to evaluate
* Most commonly used in industry

---

# Disadvantages

* Requires labelled data
* Label creation is expensive
* Data collection takes time

---

# 2. Unsupervised Learning

## Definition

In Unsupervised Learning:

```text
Input Available
Output Not Available
```

The model tries to discover hidden patterns.

---

## Example

Customer Data

| Customer | Age | Income |
| -------- | --- | ------ |
| A        | 25  | 30K    |
| B        | 27  | 35K    |
| C        | 55  | 1.2L   |

No output column exists.

The model must find patterns itself.

---

# Goal

Discover:

* Groups
* Similarities
* Patterns
* Relationships

---

# Common Tasks

## Clustering

Group similar items together.

Example:

```text
Customer Segmentation
```

Output:

```text
Group 1 → Students

Group 2 → Working Professionals

Group 3 → Senior Citizens
```

---

## Association Rule Learning

Find relationships between items.

Example:

```text
People who buy bread
often buy butter
```

Used in recommendation systems.

---

## Dimensionality Reduction

Reduce number of features while preserving information.

Used for:

* Visualization
* Faster Training
* Noise Reduction

---

# Common Algorithms

* K-Means Clustering
* Hierarchical Clustering
* DBSCAN
* PCA

---

# Real-Life Examples

### Amazon

Product Recommendations

### Marketing

Customer Segmentation

### Social Networks

Friend Suggestions

### E-commerce

Market Basket Analysis

---

# Advantages

* No labelled data needed
* Finds hidden patterns

---

# Disadvantages

* Hard to evaluate
* Results may be subjective

---

# 3. Semi-Supervised Learning

## Definition

Combination of:

* Small amount of labelled data
* Large amount of unlabelled data

---

## Example

Suppose:

100,000 Images

Labelled:

```text
1000 Images
```

Unlabelled:

```text
99,000 Images
```

Instead of manually labelling everything:

The model learns from both labelled and unlabelled data.

---

# Why Use It?

Labelling data is expensive.

Examples:

* Medical Images
* Satellite Images
* Face Recognition

---

# Real-Life Example

Facebook Photo Tagging

Initially:

Few labelled faces

Later:

System learns from unlabelled images.

---

# Advantages

* Reduces labelling cost
* Better than using small labelled dataset alone

---

# Disadvantages

* More complex
* Requires careful implementation

---

# 4. Reinforcement Learning

## Definition

The model learns by interacting with the environment.

Learning happens through:

```text
Reward
and
Penalty
```

---

# Components

## Agent

Learner

Example:

Robot

---

## Environment

World around the agent.

---

## Action

What the agent does.

Example:

Move Left

Move Right

Jump

---

## Reward

Positive feedback.

Example:

+10 Points

---

## Penalty

Negative feedback.

Example:

-5 Points

---

# Working

```text
Agent
  ↓
Take Action
  ↓
Environment
  ↓
Reward / Penalty
  ↓
Learn
```

---

# Real-Life Examples

### Self-Driving Cars

Learn driving decisions.

### Chess AI

Learns winning strategies.

### AlphaGo

Defeated world champion in Go.

### Robotics

Robot learns movement.

---

# Advantages

* Learns through experience
* Excellent for decision-making problems

---

# Disadvantages

* Requires huge training time
* Computationally expensive

---

# Quick Comparison

| Type            | Labelled Data | Main Goal         |
| --------------- | ------------- | ----------------- |
| Supervised      | Yes           | Predict Output    |
| Unsupervised    | No            | Find Patterns     |
| Semi-Supervised | Partial       | Improve Learning  |
| Reinforcement   | No Labels     | Learn via Rewards |

---

# Interview Questions

## What is Supervised Learning?

Learning from labelled data where both input and output are available.

---

## What is Unsupervised Learning?

Learning from unlabelled data to discover hidden patterns.

---

## Difference Between Classification and Regression?

Classification:

```text
Predict Categories
```

Regression:

```text
Predict Numbers
```

---

## What is Reinforcement Learning?

A learning technique where an agent learns through rewards and penalties.

---

## What is Semi-Supervised Learning?

A combination of labelled and unlabelled data used for training.

---

# Revision Sheet

```text
Supervised Learning
→ Input + Output
→ Predict Future Values
```

```text
Unsupervised Learning
→ Only Input
→ Find Hidden Patterns
```

```text
Semi-Supervised Learning
→ Few Labels
→ Many Unlabelled Samples
```

```text
Reinforcement Learning
→ Reward
→ Penalty
→ Learn Behaviour
```

The most important type in real-world industry projects is:

✅ Supervised Learning

* Classification
* Regression

Master these first before moving to Deep Learning or Reinforcement Learning.
