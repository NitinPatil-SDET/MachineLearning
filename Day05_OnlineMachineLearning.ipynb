# Introduction

In the previous lesson, we learned about **Batch (Offline) Learning**, where a model is trained once and retrained later with new data.

But what if data is generated every second?

Examples:

* Stock Market
* Social Media
* E-commerce
* Banking Transactions
* IoT Devices

Retraining the model every hour or every minute is impractical.

This is where **Online Machine Learning** comes in.

---

# What is Online Machine Learning?

**Online Machine Learning** is a learning approach where the model learns **continuously** as new data arrives.

Instead of retraining from scratch, the model updates its knowledge incrementally.

### Simple Definition

> Learn a little → Update the model → Repeat forever.

---

# Why is it called Online Learning?

The word **Online** does **not** mean internet.

It means:

* The model keeps learning while it is running.
* New data is immediately used to improve the model.

---

# How Online Learning Works

```text
Historical Data
        │
        ▼
Initial Training
        │
        ▼
Deploy Model
        │
        ▼
New Data Arrives
        │
        ▼
Update Model
        │
        ▼
Better Predictions
        │
        ▼
Repeat...
```

Unlike Batch Learning, retraining the entire model is not required after every new record.

---

# Example 1 – Stock Market Prediction

Suppose a model predicts stock prices.

Every second:

* New prices arrive
* Trading volume changes
* Market sentiment changes

Instead of waiting one month to retrain:

```text
New Data
     ↓
Model Updates
     ↓
New Prediction
```

The model keeps adapting.

---

# Example 2 – Spam Detection

Initially:

```text
Spam Email
```

The model classifies correctly.

After some time:

Spammers invent new tricks.

Example:

```text
WIN MONEY
```

becomes

```text
W!N M0NEY
```

If the model is using Online Learning, it gradually learns these new patterns without full retraining.

---

# Online Learning Workflow

```text
Collect Data
      │
      ▼
Train Initial Model
      │
      ▼
Deploy
      │
      ▼
Receive New Data
      │
      ▼
Update Model
      │
      ▼
Improve Accuracy
      │
      ▼
Continue Learning
```

---

# What is Incremental Learning?

Online Learning is also called **Incremental Learning**.

Reason:

The model learns in **small increments** instead of learning everything again.

Think of it like adding one new chapter to your knowledge every day instead of rereading the entire book.

---

# Mini-Batch Learning

Sometimes data arrives in groups.

Instead of learning from:

```text
1 record
```

the model learns from:

```text
100 records
```

or

```text
1000 records
```

This approach is called **Mini-Batch Learning**.

It is a balance between Batch Learning and pure Online Learning.

---

# Advantages of Online Learning

## Learns Continuously

The model improves whenever new data arrives.

---

## Handles Large Data

No need to store the complete dataset in memory.

Suitable for massive datasets.

---

## Adapts to Changing Patterns

Useful when data changes frequently.

Examples:

* Customer behaviour
* Fraud patterns
* Market prices

---

## Lower Memory Usage

Only a small portion of data is processed at a time.

---

## Faster Updates

No need to retrain from scratch.

---

# Disadvantages

## Sensitive to Bad Data

If incorrect data enters the system:

The model may learn wrong patterns.

---

## Harder to Debug

Since the model changes continuously, reproducing old behaviour can be difficult.

---

## Requires Monitoring

Continuous learning requires continuous monitoring to ensure quality.

# When Should You Use Online Learning?

Use Online Learning when:

* Data arrives continuously.
* Predictions must stay up to date.
* Full retraining is expensive.
* Data is too large to fit into memory.

---

# When Should You Avoid Online Learning?

Avoid Online Learning when:

* Data rarely changes.
* The dataset is small.
* Explainability and stability are more important than continuous updates.

In these situations, Batch Learning is often the better choice.

---

# Online Learning vs Batch Learning

| Batch Learning         | Online Learning                  |
| ---------------------- | -------------------------------- |
| Trains once            | Learns continuously              |
| Uses complete dataset  | Uses one record or small batches |
| Requires retraining    | Updates incrementally            |
| Better for static data | Better for dynamic data          |
| Higher retraining cost | Lower incremental update cost    |
| Simpler pipeline       | More complex pipeline            |

# Batch vs Online Flow

# Common Misconceptions

❌ **Online Learning means internet-based learning.**

✔ No. It means the model learns continuously after deployment.

---

❌ **Online Learning always learns from one record at a time.**

✔ Not necessarily. It can also learn from small batches (Mini-Batch Learning).

---

# Actionable Insights

* Ask this question before choosing a learning strategy:

> **Will my data change frequently after deployment?**

* If **No**, Batch Learning is usually enough.
* If **Yes**, Online Learning is often the better choice.

# Glossary

| Term                     | Meaning                                                                      |
| ------------------------ | ---------------------------------------------------------------------------- |
| **Online Learning**      | A learning approach where the model updates continuously after deployment.   |
| **Incremental Learning** | Another name for Online Learning because knowledge is added gradually.       |
| **Mini-Batch Learning**  | Updating the model using small groups of data instead of the entire dataset. |
| **Streaming Data**       | Data that arrives continuously over time.                                    |
| **Model Update**         | Adjusting the model using newly available data.                              |

# Key Takeaway

Online Machine Learning is designed for environments where **data changes continuously**. Instead of rebuilding the model from scratch, it **learns incrementally**, making it suitable for real-time systems such as fraud detection, recommendation engines, online advertising, and financial markets. The trade-off is that continuous learning requires careful monitoring to prevent poor-quality data from degrading the model.
