# Big Picture

Many beginners think AI, ML, and DL are separate technologies.

They are actually related like this:

```text
Artificial Intelligence (AI)
│
└── Machine Learning (ML)
     │
     └── Deep Learning (DL)
```

Think of it like:

```text
Vehicle
│
├── Car
│
└── Sports Car
```

- Vehicle = AI
- Car = ML
- Sports Car = DL

---

# What is Artificial Intelligence?

Artificial Intelligence is the ability of a machine to perform tasks that normally require human intelligence.

Examples:

- ChatGPT
- Self-driving cars
- Face recognition
- Google Maps
- Chess-playing computers
- Recommendation systems

### Goal of AI

```text
Make machines behave intelligently.
```

---

# Traditional AI (Rule-Based Systems)

Early AI systems were built using rules.

Example:

```text
IF fever = yes
AND cough = yes

THEN flu
```

Everything depended on manually written rules.

---

## Problems with Rule-Based AI

Imagine creating a self-driving car.

You would need millions of rules:

```text
IF traffic signal = red
THEN stop

IF pedestrian detected
THEN stop

IF road clear
THEN move
```

Problems:

- Difficult to scale
- Difficult to maintain
- Cannot learn
- Cannot adapt to new situations

This led to Machine Learning.

---

# What is Machine Learning?

Machine Learning is a subset of AI that allows systems to learn patterns from data.

Instead of writing rules manually:

```text
Data → Learn Pattern → Prediction
```

---

## Traditional Programming

```text
Input + Rules
       ↓
     Output
```

Example:

```text
Marks = 95

Rule:
Marks > 40

Output:
Pass
```

---

## Machine Learning

```text
Input + Output
       ↓
 Machine Learns Rules
```

Example:

Training Data:

```text
Study Hours → Pass
Study Hours → Fail
Study Hours → Pass
```

The model automatically learns the relationship.

---

# Definition of Machine Learning

A computer program is said to learn if its performance improves with experience.

### Core Idea

```text
Learning from Data
```

# What is a Feature?

A Feature is an input variable used by the model.

Example:

House Price Prediction

Features:

- Area
- Bedrooms
- Location
- Age of House

---

# What is Feature Engineering?

Feature Engineering means converting raw data into useful information for the model.

Example:

Raw Data:

```text
Date of Birth
```

Feature:

```text
Age
```

Another Example:

Raw Image

Convert into:

- Edges
- Shapes
- Textures

---

# Why Feature Engineering Matters

Good Features:

```text
Good Model
```

Bad Features:

```text
Bad Model
```

Even the best algorithm performs poorly if the features are poor.

---

# What is Deep Learning?

Deep Learning is a subset of Machine Learning.

It uses Neural Networks with many layers.

---

# Hierarchy

```text
AI
│
└── ML
     │
     └── DL
```

---

# Biggest Difference

Machine Learning:

```text
Human creates features
```

Deep Learning:

```text
Model learns features automatically
```

This is the biggest advantage of Deep Learning.

---

# Neural Networks

Deep Learning is based on Artificial Neural Networks.

Inspired by the human brain.

Structure:

```text
Input Layer
      ↓
Hidden Layer
      ↓
Hidden Layer
      ↓
Output Layer
```

---

# What is a Perceptron?

A Perceptron is the smallest unit of a Neural Network.

Example:

Inputs:

- Study Hours
- Attendance

Output:

- Pass
- Fail

Thousands or millions of perceptrons form a Neural Network.

---

# Face Recognition Example

## Layer 1

Learns:

- Lines
- Edges

---

## Layer 2

Learns:

- Shapes
- Curves

---

## Layer 3

Learns:

- Eyes
- Nose
- Mouth

---

## Final Layer

Learns:

```text
Human Face
```

No manual feature engineering is required.

# Machine Learning vs Deep Learning

| Machine Learning | Deep Learning |
|------------------|--------------|
| Requires Feature Engineering | Learns Features Automatically |
| Works with Small Data | Requires Large Data |
| Less Computing Power | More Computing Power |
| Easier to Explain | Harder to Explain |
| Faster Training | Slower Training |
| Best for Structured Data | Best for Unstructured Data |

---

# When to Use Machine Learning

Use ML when:

- Data is structured
- Dataset is small
- Explainability is important
- Computing resources are limited

Examples:

- Banking
- Insurance
- Customer Churn Prediction
- Fraud Detection

---

# When to Use Deep Learning

Use DL when:

- Large dataset available
- Image processing
- Audio processing
- Video processing
- NLP tasks

Examples:

- ChatGPT
- Face Recognition
- Speech Recognition
- Self-Driving Cars

# Interview Questions

## What is AI?

Artificial Intelligence is the field of building systems capable of performing tasks that normally require human intelligence.

---

## What is Machine Learning?

Machine Learning is a subset of AI that learns patterns from data without explicit programming.

---

## What is Deep Learning?

Deep Learning is a subset of Machine Learning that uses multi-layer neural networks to learn complex patterns.

---

## Why is Deep Learning better for images?

Because it automatically learns features such as edges, shapes, and objects from raw images.

---

## What is Feature Engineering?

Feature Engineering is the process of converting raw data into meaningful features for machine learning models.

---

# Quick Revision

```text
AI = Intelligence

ML = Learning from Data

DL = Learning using Deep Neural Networks
```

```text
Traditional AI → Rules

Machine Learning → Learn Rules

Deep Learning → Learn Rules + Learn Features
```

---

# Key Takeaway

Artificial Intelligence is the goal.

Machine Learning is one way to achieve that goal using data.

Deep Learning is an advanced form of Machine Learning that automatically learns features and patterns using neural networks.
